Import libraries

```python
import yfinance as yf
import pandas as pd
import os
```

```python
# Define which tickers we want and the date range
tickers = ["SPY", "QQQ", "AAPL"]
start_date = "2014-01-01"
end_date = "2024-01-01"

# Download the data
data = yf.download(tickers, start=start_date, end=end_date)

# Preview the first 5 rows
data.head()
```

```python
# Pull out just SPY's data
spy = data["Close"]["SPY"]

# Preview it
spy.head()
```

```python
# Get all OHLCV columns for SPY
spy = data.xs("SPY", axis=1, level=1)

# Preview it
spy.head()
```

```python
# Create the raw data folder if it doesn't exist
os.makedirs("../data/raw", exist_ok=True)

# Save SPY data as a CSV
spy.to_csv("../data/raw/SPY.csv")

print("Saved!")
```

```python
# Load SPY data from CSV instead of downloading
spy_loaded = pd.read_csv("../data/raw/SPY.csv", index_col="Date", parse_dates=True)

spy_loaded.head()
```

```python
# Save all tickers to CSV
tickers = ["SPY", "QQQ", "AAPL"]

for ticker in tickers:
    ticker_data = data.xs(ticker, axis=1, level=1)
    ticker_data.to_csv(f"../data/raw/{ticker}.csv")
    print(f"Saved {ticker}")
```
